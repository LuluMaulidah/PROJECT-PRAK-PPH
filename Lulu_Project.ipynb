## Data exploration

### Subtask:
Explore the loaded dataframes (df_transaksi, df_aset, df_kebijakan) to understand their structure, data types, and potential relationships between columns. Identify key variables and potential trends within each dataset.


**Reasoning**:
Explore the dataframes to understand their structure, data types, potential relationships between columns, and potential trends.  This involves displaying the shape, data types, descriptive statistics, missing values, unique values in categorical columns, potential relationships (correlation for numerical, distribution for categorical), and trends (especially over time if a time variable exists). Visualizations will help to understand the distribution of key numerical variables.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to analyze a dataframe
def analyze_dataframe(df, df_name):
    print(f"Analysis of DataFrame: {df_name}")
    print(f"Shape: {df.shape}")
    print("\nData Types:")
    print(df.dtypes)
    print("\nDescriptive Statistics:")
    print(df.describe(include='all'))
    print("\nMissing Values:")
    missing_percentage = (df.isnull().sum() / len(df)) * 100
    print(missing_percentage)
    print("\nUnique Values in Categorical Columns:")

    for col in df.columns:
        if df[col].dtype == 'object':
          print(f"Column '{col}': {df[col].unique()}")

    # Numerical columns correlation
    numerical_cols = df.select_dtypes(include=['number']).columns
    if len(numerical_cols) > 1:
        print("\nCorrelation Matrix for Numerical Columns:")
        print(df[numerical_cols].corr())

    # Visualizations for numerical columns
    for col in numerical_cols:
      plt.figure(figsize=(8, 6))
      plt.hist(df[col], bins=10, edgecolor='black')
      plt.title(f'Distribution of {col}')
      plt.xlabel(col)
      plt.ylabel('Frequency')
      plt.show()

# Analyze each dataframe
analyze_dataframe(df_transaksi, 'df_transaksi')
analyze_dataframe(df_aset, 'df_aset')
analyze_dataframe(df_kebijakan, 'df_kebijakan')
```

```python
# Fill missing values in 'tax_holiday_awal' and 'tax_holiday_akhir' with 0.
# Rationale:  Imputing with 0 indicates the absence of a tax holiday in those years, which is a valid value.
# Correcting FutureWarning by not using inplace=True
df_kebijakan['tax_holiday_awal'] = df_kebijakan['tax_holiday_awal'].fillna(0)
df_kebijakan['tax_holiday_akhir'] = df_kebijakan['tax_holiday_akhir'].fillna(0)


# Assuming 'tahun' column is missing in df_aset and the index represents the year.
# Add it based on the index, assuming the index corresponds to years starting from a certain point.
# If your df_aset index does not represent the year, you will need to adjust this logic.
# For example, if the index starts from 0 and the first year is 2020, use df_aset.index + 2020.
# Based on the success of the previous block, we will keep this approach.
if 'tahun' not in df_aset.columns:
    # Add a 'tahun' column assuming the index starts from 0 and represents years from 2020
    # You might need to adjust '2020' based on your actual data.
    # Check the index of df_aset to confirm if this assumption is correct.
    df_aset['tahun'] = df_aset.index + 2020


# Convert 'tahun' columns to datetime objects for easier merging and analysis
# This step assumes the 'tahun' column represents the year.
df_transaksi['tahun'] = pd.to_datetime(df_transaksi['tahun'], format='%Y')
df_aset['tahun'] = pd.to_datetime(df_aset['tahun'], format='%Y')
df_kebijakan['tahun'] = pd.to_datetime(df_kebijakan['tahun'], format='%Y')

# Merge dataframes using 'tahun' as the key.  Use an inner join.
# Rationale: An inner join ensures that only years present in all three datasets are included.
df_merged = pd.merge(df_transaksi, df_kebijakan, on='tahun', how='inner')
df_merged = pd.merge(df_merged, df_aset, on='tahun', how='inner')

# Feature Engineering: Calculate profit margin.
# Ensure that 'pendapatan' is not zero to avoid division by zero.
df_merged['profit_margin'] = ((df_merged['pendapatan'] - df_merged['beban_operasional'] - df_merged['penyusutan']) / df_merged['pendapatan'].replace(0, pd.NA)) * 100
# Fill resulting NaNs from division by zero or original NaNs in components if necessary
df_merged['profit_margin'] = df_merged['profit_margin'].fillna(0) # Or another appropriate value


display(df_merged.head())
```

## Data visualization

### Subtask:
Visualize relationships and trends in the merged dataframe.


**Reasoning**:
Create visualizations to explore the relationships between different variables in the merged dataframe, focusing on profit margin and its relation to other key variables, while differentiating by scenario.



```python
import matplotlib.pyplot as plt

# Plot 1: Profit Margin vs. Tax Rate
plt.figure(figsize=(10, 6))
for scenario in df_merged['skenario'].unique():
    subset = df_merged[df_merged['skenario'] == scenario]
    plt.scatter(subset['tax_rate'], subset['profit_margin'], label=scenario, alpha=0.7)
plt.xlabel('Tax Rate')
plt.ylabel('Profit Margin (%)')
plt.title('Profit Margin vs. Tax Rate by Scenario')
plt.legend()
plt.grid(True)
plt.show()


# Plot 2: Profit Margin vs. Pendapatan
plt.figure(figsize=(10, 6))
for scenario in df_merged['skenario'].unique():
    subset = df_merged[df_merged['skenario'] == scenario]
    plt.scatter(subset['pendapatan'], subset['profit_margin'], label=scenario, alpha=0.7)
plt.xlabel('Pendapatan')
plt.ylabel('Profit Margin (%)')
plt.title('Profit Margin vs. Pendapatan by Scenario')
plt.legend()
plt.grid(True)
plt.show()

# Plot 3: Profit Margin vs. Nilai Perolehan
plt.figure(figsize=(10, 6))
for scenario in df_merged['skenario'].unique():
    subset = df_merged[df_merged['skenario'] == scenario]
    plt.scatter(subset['nilai_perolehan'], subset['profit_margin'], label=scenario, alpha=0.7)
plt.xlabel('Nilai Perolehan')
plt.ylabel('Profit Margin (%)')
plt.title('Profit Margin vs. Nilai Perolehan by Scenario')
plt.legend()
plt.grid(True)
plt.show()
```
